```python
# import dependencies
import pandas as pd
import plotly.express as px
import seaborn as sns
from dash import Dash, html, dcc
```

```python
top_songs = pd.read_csv('top_10000_1960-now.csv')
top_songs.head()
```

```python
# Cleaning the data

# Dropping unecessary columns

top_songs.drop(columns=['Track URI', 'Artist URI(s)', 'Album URI', 
                        'Album Artist URI(s)', 'Track Number', 'Disc Number' , 
                        'Track Preview URL', 'Speechiness', 'Album Artist Name(s)',
                        'Acousticness', 'Instrumentalness', 'Valence', 
                        'Tempo', 'Time Signature', 'Album Genres', 
                        'Label', 'Copyrights', 'Album Image URL', 
                        'ISRC', 'Added By', 'Added At', 'Energy',
                        'Key', 'Loudness', 'Mode', 'Liveness', 'Danceability', 'Popularity'], inplace=True) 

# Track URI, Artist URI(s), Album URI, Album Artist URI(s), Track Preview URL, and Album Image URL dropped because it provides no data to users
    # May add URI's later if I decide to link data points directly to spotify
# Track number and Disc number dropped because I do not believe users will care about this information
# Album Artist Name(s) dropped because both are already provided in other columns
# 'Acousticness', 'Instrumentalness', 'Valence', 'Tempo', 'Time Signature', 'Energy', 'Key', 'Loudness', 'Mode', 'Liveness', 'Danceability' dropped because most of these are subjective, not concrete, and not very useful to users.
    # The ones that aren't subjective such as tempo, time signiture, and key, are just not what users would be interested in when looking for songs
#'Added By', 'Added At' dropped because this data is not useful to users
# Label', 'Copyrights', 'ISRC' dropped because this data is unrelated to the tracks themselves and users can find themselves if they are interested in the songs. Cannot really make useful 
# Album Genre dropped because already have artist genre which I believe is more useful
# Popularity dropped because the vast majority of tracks do not have values

top_songs.head()
```

```python
# Convert track duration from miliseconds to seconds
top_songs['Track Duration (ms)'] = top_songs['Track Duration (ms)'] / 1000
top_songs.rename(columns={'Track Duration (ms)': 'Track Duration (s)'}, inplace=True)
top_songs.head()
```

```python
top_songs.dropna(subset=['Track Name'], inplace=True) # 1 row with missing data
top_songs.dropna(subset=['Artist Name(s)'], inplace=True) # 1 row with missing data
top_songs.dropna(subset=['Album Name'], inplace=True) # 1 row with missing data
top_songs.dropna(subset=['Explicit'], inplace=True) # 2 rows with missing data
top_songs.dropna(subset=['Artist Genres'], inplace=True) # 550 rows with missing data
```

```python
top_songs.to_csv('data.csv') # retreive csv file of cleaned data
```
